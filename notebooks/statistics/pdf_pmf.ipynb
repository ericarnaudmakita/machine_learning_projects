* Probability Density Function (use for continous data)

For a given range of values it tells us the Probability of that range occuring





* Probability Mass Function (use for discrete data)

Probability of discrete data occuring.

Example of Data Distribution

1 - Uniform Distribution

There is flat constant probability occuring of a vlaue  within given  range

```python
import numpy as np
import matplotlib.pyplot as plt

value = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(value, 50)
plt.show()
```

2 - Normal / Gaussian Distribution

Visualize the probability densisity function

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

Generate some random numbers with a normal distribution. "mu" is the desired mean, "sigma" is the std

```python
mu = 5.0
sigma = 2.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

Exponential PDF / "Power Law"

```python
from scipy.stats import expon

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

Binomial Probability Mass Function

```python
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

Poisson Probability Mass Function

Example: My website gets on average 500 visits per day. What's the odds of getting 550?

```python
from scipy.stats import poisson

mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
