* Probability Density Function (use for continous data)

For a given range of values it tells us the Probability of that range occuring





* Probability Mass Function (use for discrete data)

Probability of discrete data occuring.

Example of Data Distribution

1 - Uniform Distribution

There is flat constant probability occuring of a vlaue  within given  range

```python
import numpy as np
import matplotlib.pyplot as plt

value = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(value, 50)
plt.show()
```

2 - Normal / Gaussian Distribution

Visualize the probability densisity function

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```
