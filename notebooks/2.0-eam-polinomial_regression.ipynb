```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import pandas as pd
import seaborn as sns
sns.set()

import plotly.express as px
```

Not all relationships are linear

* Second order polynomial : $$ y = a{x}^2 + bx + c$$

We get the values of a, b and c from the polynomial regression ...

* Third order : $$ y = a{x}^3 + b{x}^2 + cx + d$$

* Higher orders produce more complex curves.

Beware overfitting


* Don't use more degrees than we need

* We need to visualize our data first to see how complex of a curve there might really be

* We need to visualize the fit - is our curve going out of its way to accomadate outliers?

* A high `r-squared` simply means our curve futs our training data well; but it may not be a good predictor.



What if our data doesn't look linear at all?

```python
np.random.seed(2)
pageSpeeds =  np.random.normal(3.0, 1.0, 1000)
purchaseAmout = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

plt.scatter(pageSpeeds, purchaseAmout);
```

numpy has a handy polyfit function we can use, to let us construct an nth-degree polynomial model of our data that minimizes squared error.

Let's try it with a 4th degree polynomial

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmout)

p4 = np.poly1d(np.polyfit(x, y, 4))
```

We'll visualize our original scatter plot, together with a plot of our predicted values using the polynomial for page speed times ranging from 0_7 seconds:

```python
xp = np.linspace(0, 7, 100)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

Look pretty good!, we might ask ourself if we overfitting. Let's measure the r-squared error:

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p4(x)) # compare a set of observations with a set of predictions

print(r2)
```
