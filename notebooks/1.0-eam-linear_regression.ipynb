Linear Regression

* Fit a line to a data set of observations

* Use this line to predict unobesrved values. We can use it to predict points in the future, the past, whatever. In fact time usually has nothing to do with it.

* Usually using `least squares`


Measuring error witj r-squared

* How do we measure how well our line fits our data?

* R-squared (aka coefficient of determination) measures:

The fraction of the total variation in Y that is captured by the model (or how well our line follows the variations that happening)

$$ 1.0 - \frac{sum \space of \space squared \space errors}{sum \space of \space squared \space variation \space from \space mean}$$

Interpreting r-squared: 

Quantutative measure of how good a given regression is to a set of data points and use that to chose the model that best fit our data

* Ranges from 0 to 1

* 0 is bad (none of the variance is captured), 1 is good (all of the variance is captured) => low r-squared is poor fit and higher r-squared is perfect fit

Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```python
pageSpeeds =  np.random.normal(3.0, 1.0, 1000)
purchaseAmout = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

plt.scatter(pageSpeeds, purchaseAmout);
```

As we only have two features, we can keep it simple and just use scipy.state.linregress:

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmout)
```

Not surprinsgly, our R-squared value shows a really good fit:

```python
r_value ** 2
```

Let's use the slope and intercept we got from the regression to plot predicted values vs. observed:

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmout)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```
